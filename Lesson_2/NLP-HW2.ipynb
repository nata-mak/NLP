```python
PATH = '/Users/admin/Desktop/GeekBrains/NLP/2-Создание признакового пространства/corpus.txt'
```

```python
# Загружаем данные
data = open(PATH).read()
```

```python
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
df = pd.DataFrame()
df['text'] = texts
df['label'] = labels
df.head(5)
```

```python
df.info()
```

```python
df['label'].value_counts()
```

### Предобработка текста

```python
# замена апострофов
apostrophe_dict = {"ain't": "am not", "can't":"can not", "isn't": "is not", "don't": "do not", 
                 "doesn't": "does not", "won't": "will not",
                  "aren't": "are not", "wasn't": "was not", "didn't": "did not", "I'm":"I am",
                  "hasn't": "has not"}

def replace_apostrophe(text):
    return " ".join(apostrophe_dict.get(x, x) for x in text.split())

df['text'] = df['text'].apply(replace_apostrophe)
```

```python
def remove_punct(string):
    punct_pattern = re.compile('[^\w\s]+')
    return punct_pattern.sub(r' ', string)

df['text'] = df['text'].apply(remove_punct)
```

```python
def remove_symb(string):
    punct_pattern = re.compile('[^a-zA-Z0-9]')
    return punct_pattern.sub(r' ', string)

df['text'] = df['text'].apply(remove_symb)
```

```python
def remove_digit(string):
    punct_pattern = re.compile('[^a-zA-Z]')
    return punct_pattern.sub(r' ', string)

df['text'] = df['text'].apply(remove_digit)
```

```python
def del_short_word(text):
    return " ".join([w for w in text.split() if len(w)>1])

df['text'] = df['text'].apply(del_short_word)
```

### Инициализация векторайзеров

```python
count_vectorizer_1 = CountVectorizer(ngram_range=(1, 1), max_df=0.9,  max_features=1000, stop_words='english')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.25, random_state=42)
```

```python
# labelEncode для целевой переменной переведем в бинарный формат
encoder = preprocessing.LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
bow_1 = count_vectorizer_1.fit_transform(X_train) #(ngram_range=(1, 1)
# lr_model = LinearRegression(random_state=42)
gb_model = GradientBoostingRegressor(max_depth=5,
                                     min_samples_leaf=7,
                                     random_state=42,  
                                     n_estimators=100)
```

```python
# списки для сохранения результатов
vect_method = ['Count_vectorizer_ngram_(1,1)', 'Count_vectorizer_ngram_(1,2)', 'Count_vectorizer_ngram_(2,2)',
               'Count_vectorizer_ngram_(2,3)','Tfidf']
score = []
```

```python
gb_model.fit(bow_1, y_train)
```

```python
pred = gb_model.predict(count_vectorizer_1.transform(X_test))
print(classification_report(pred.round(), y_test))
```

```python
from sklearn import metrics
f1_score_1 = metrics.f1_score(pred.round(), y_test)
score.append(f1_score_1)
f1_score_1
```

### Снижение размерности с помощью PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10, random_state=42)
```

```python
feature_names = count_vectorizer_1.get_feature_names()
X_train_lemm = pd.DataFrame(bow_1.toarray(), columns = feature_names)
X_train_lemm
```

```python
bow_1_test = count_vectorizer_1.transform(X_test)
X_test_lemm = pd.DataFrame(bow_1_test.toarray(), columns = feature_names)
X_test_lemm.head()
```

```python
X_train_pca = pca.fit_transform(X_train_lemm)
len(X_train_pca)
```

```python
X_test_pca = pca.transform(X_test_lemm)
X_test_pca
```

```python
gb_model.fit(X_train_pca, y_train)
pred_pca = gb_model.predict(X_test_pca)
print(classification_report(pred_pca.round(), y_test))
```

```python
f1_score_pca = metrics.f1_score(pred_pca.round(), y_test)
print(round(f1_score_pca, 3))
```

Сокращение размерности модели методом PCA до 10 компонент привело к довольно неплохому качеству по метрике F1-score. Зато число фичей в разы уменьшилось с 1000 до 10 (по сравнению с CountVectorizer).

```python
# попробуем снизить размерность до 500 компонент:
pca_50 = PCA(n_components=50, random_state=42)
X_train_pca_50 = pca_50.fit_transform(X_train_lemm)
X_test_pca_50 = pca_50.transform(X_test_lemm)
```

```python
gb_model.fit(X_train_pca_50, y_train)
pred_pca_50 = gb_model.predict(X_test_pca_50)
print(classification_report(pred_pca_50.round(), y_test))
```

В результате снжения размерности в 20 раз (с 1000 до 50 фичей) получили отличное качество у модели, сравнимое с tdidf-векторайзером на 1000 признаков.
